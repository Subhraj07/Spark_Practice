###### Processing using NaiveBayesAnalyzer

```python
sentimentcheck = []
def classify():
    length = len(trip1)
    trans = []
    for i in range(length):
        data = tb(trip1["Review Description"][i])
        if data.detect_language() == "en":
            trans.append(data)
            continue
        else:
            trans.append(data.translate(to='en')) 
    for text in trans:
        sentimentcheck.append(text.sentiment)
```

###### Processing using TextBlob

```python
normalsentimentcheck = []
def normalsentiment():
    length = len(trip1)
    trans = []
    for i in range(length):
        data = TextBlob(trip1["Review Description"][i])
        if data.detect_language() == "en":
            trans.append(data)
            continue
        else:
            trans.append(data.translate(to='en'))    
    for text in trans:
        normalsentimentcheck.append(text.sentiment)       
```

```python
norpol = []
norsub = []
norlen = len(normalsentimentcheck)
for nor in range(norlen):
    norpol.append(normalsentimentcheck[nor].polarity)
    norsub.append(normalsentimentcheck[nor].subjectivity)
```

```python
trip1["polarity"] = norpol
trip1["subjectivity"] = norsub
```

```python
trip1.to_excel("./trip1.xlsx")
```

```python
trip2 = pd.read_excel("TripAdvisor Example 2.xlsx")
normalsentimentcheck2 = []
def normalsentiment2():
    length = len(trip2)
    trans = []
    for i in range(length):
        data = TextBlob(trip2["Review Text"][i])
        if data.detect_language() == "en":
            trans.append(data)
            continue
        else:
            trans.append(data.translate(to='en'))    
    for text in trans:
        normalsentimentcheck2.append(text.sentiment) 
```

```python
start_time = time.time()
normalsentiment2()
print("--- %s seconds ---" % (time.time() - start_time))
```
