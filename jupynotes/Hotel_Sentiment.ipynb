```python
from textblob import TextBlob
import pandas as pd
import time
pd.set_option('display.max_colwidth', -1)
# pd.options.display.max_colwidth = 10000
```

```python
def normalsentiment(hotel):
    length = len(hotel)
    trans = []
    for i in range(length):
        data = TextBlob(hotel["review"][i])
        if data.detect_language() == "en":
            trans.append(data)
            continue
        else:
            trans.append(data.translate(to='en'))   
    return trans        
```

### Embassy_Suits

```python
start_time = time.time()
embassy_suits = pd.read_excel("./hotel_details/Embassy_Suites.xlsx")

embassy_suits_sentiment = []
embassy_suits_trans = normalsentiment(embassy_suits) 
embassy_suits_pol = []
embassy_suits_sub = []


for text in embassy_suits_trans:
        embassy_suits_sentiment.append(text.sentiment)
        
len_embassy_suits_sentiment = len(embassy_suits_sentiment)        
        
for nor in range(len_embassy_suits_sentiment):
    embassy_suits_pol.append(embassy_suits_sentiment[nor].polarity)
    embassy_suits_sub.append(embassy_suits_sentiment[nor].subjectivity)
    
embassy_suits["polarity"] = embassy_suits_pol
embassy_suits["subjectivity"] = embassy_suits_sub    

embassy_suits.to_excel("./polarity_hotel/embassy_suits.xlsx")
embassy_suits.sort_values(['polarity'], ascending=[True])[:11].to_excel("./polarity_hotel/neg_embassy_suits.xlsx")
embassy_suits.sort_values(['polarity'], ascending=[False])[:11].to_excel("./polarity_hotel/pos_embassy_suits.xlsx")

print("Completed in seconds" , time.time() - start_time)
```
